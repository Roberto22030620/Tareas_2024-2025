```python

import matplotlib.pyplot as plt
import numpy as np
import random

puntos=[]
for i in range(50):
    puntos.append([random.randint(1,10), random.randint(1,10)])

puntos=np.array(puntos)
#plt.scatter(puntos[:,0], puntos[:,1])
centrox=random.randint(1,10)
centroy=random.randint(1,10)
plt.scatter(centrox, centroy,marker='X', color='red')

for i in range(50):
    distancia=np.sqrt((puntos[i,0]-centrox)**2+(puntos[i,1]-centroy)**2)
    #print(distancia)
    if distancia<2:
        plt.scatter(puntos[i,0], puntos[i,1], color='blue')
    else:
        plt.scatter(puntos[i,0], puntos[i,1], color='green')

circle=plt.Circle((centrox,centroy),2,fill=False, color='red')
plt.gca().add_artist(circle)
plt.axis('equal')

plt.show()
```
